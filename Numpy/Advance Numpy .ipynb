## SPLIT NUMPY ARRAY

```python
import numpy as np
```

```python
p=np.arange(8)
p
```

```python
np.split(p,2)# Split the numpy array into two equal subset
```

```python
np.split(p,5)# Split will throw error if len(p) not divisible by 5
```

```python
np.split(p,[3,6])# Split array using the index list the will create len(indexList)+1 element
```

```python
companyData = np.array([('IBM', 'Apple Inc.', 'Intel', 'Sony', 'Dell', 'Akamai'),
                 ('New York', 'California', 'California', 'Texas', 'Washington', 'Massachusetts')])
```

```python
print("Hsplit divide the array columnWise into parts")
print("Alternate to Hsplit is use split with axis 1")
```

```python
np.hsplit(companyData,2)#Divide the 2D numpy array into two parts columnwise using hsplit
```

```python
np.split(companyData,2,axis=1)#Divide the 2D numpy array into two parts columnwise using split
```

```python
print("Vsplit divide the array Rowwise into parts")
print("Alternate to Hsplit is use split with axis 0")
```

```python
np.vsplit(companyData,2)#Divide the 2D numpy array into two parts rowise using vsplit
```

```python
np.split(companyData,2,axis=0)#Divide the 2D numpy array into two parts rowise using split
```

## Making shallow copy using view 

```python
countries = np.array(['India', 'Peru', 'USA', 'Nigeria', 'Pakistan', 'Germany'])
```

```python
loc1=countries.view()
loc2=countries.view()
```

```python
print('Countries id:',id(countries),countries)
print('view1:',id(loc1),loc1)
print('view2:',id(loc2),loc2)     
```

```python
print("Thier values are identical but thier ids are different ")
loc1[2]="UK"
```

```python
print('Countries:',countries)
print('view1:',loc1)
print('view2:',loc2)     
```

```python
loc1.shape=(2,3)
print('Countries:',countries)
print('view1:',loc1)
print('view2:',loc2)
print("change in Shape is not reflected in the original object")
```

```python
loc2=np.array(["ind","peru","USA"])
```

```python
print('Countries:',countries)
print('view1:',loc1)
print('view2:',loc2)     
```

## Deep Copy 

```python
countries = np.array(['India', 'Peru', 'USA', 'Nigeria', 'Pakistan', 'Germany'])
```

```python
loc1=countries.copy()
loc1[2]="UK"
```

```python
print('Countries:',countries)
print('Copy',loc1)#They are indepenent of each other there is no link between them so change in the copy did not
#reflected in the original object
```

## Masking 

```python
sqArr=np.arange(16)**2
sqArr[3],sqArr[7],sqArr[9],sqArr[15]
```

```python
mask1=[3,7,9,15]
sqArr[mask1]
```

```python
mask2=np.array([[3,7],[9,15]])#Apply 2D mask  in 1D
sqArr[mask2]
```

```python
cities = np.array([['Sydney', 'Auckland', 'Vancouver'],
                  ['Lima', 'Bogota', 'Lisbon'],
                  ['Istanbul', 'Tehran', 'Colombo']])
```

```python
row = np.array([[0,0], [2,2]])
col = np.array([[0,2], [0,0]])
cities[row,col]
```

```python
cities[2,1]# indexing single element
```

```python
cities[row,col]=""# Modify array is masking
cities
```

## Advance Masking Numpy array

```python
import pandas as pd
```

```python
df=pd.read_csv("countries_of_the_world.csv",sep=',')
df.head(7)
```

```python
birthrate=df["Birthrate"].apply(lambda x:float(str( x).replace(',','.')))
```

```python
BirthRateWOna=birthrate[~pd.isna(birthrate)]
```

```python
print("Mean:",BirthRateWOna.mean())
print("Median:",BirthRateWOna.median())
```

```python
plt.plot(BirthRateWOna)
```

```python
np.count_nonzero(BirthRateWOna[BirthRateWOna>20])
```

```python
np.sort(BirthRateWOna)[:10]
```

## Masking using Boolean

```python
x=np.arange(12).reshape(4,-1,order='F')
x
```

```python
x>2# return a boolean array of same shape as x
```

```python
x[x>2]# returns a 1 D array
```

```python
np.count_nonzero(x>6)# false-->0 True-->1 so result is count of all Trues
```

```python
np.count_nonzero(x>6,axis=1)# non zero in each row
```

```python
np.any(x>10)#to check if any value is True
```

```python
np.all(x>10)# to check if all vlue is True
```
